### Problem 2 (50 points) ,
* (10 points) Find the point in the plane $x_1+2x_2+3x_3=1$ in $\\mathbb{R}^3$ that is nearest to the point $(-1,0,1)^T$. Is this a convex problem? Hint: Convert the problem into an unconstrained problem using $x_1+2x_2+3x_3=1$.
* (40 points) Implement the gradient descent and Newton's algorithm for solving the problem. Attach your codes along with a short summary including (1) the initial points tested, (2) corresponding solutions, (3) a log-linear convergence plot.

```python
import numpy as np
import matplotlib.pyplot as plt

def obj(x):
    obj = (2-2*x[0]-3*x[1])**2 + (x[0])**2 + (x[1]-1)**2
    return obj

def grad(x):
    grad = [-8+10*x[0]+12*x[1], -14+12*x[0]+20*x[1]]
    return np.squeeze(np.array(grad))

# transpose of grad(x)
def gradT(x):
    grad = [-8+10*x[0]+12*x[1]], [-14+12*x[0]+20*x[1]]
    return np.squeeze(grad)

eps = 0.001
x0 = [0,0]
k = 0
k_for_plot = [0]
soln = [x0]
x = soln[k]
error = np.linalg.norm(grad(x))
error_for_plot = [error]

# Armijo Line Search
def line_search(x):
    a = 0.06
    d = grad(x)*-1
    def phi(a,x):
        phi = obj(x)-a*0.8*np.dot(gradT(x),d)
        return phi
    while phi(a,x) < obj(x-a*grad(x)):
        a = a*0.5
    return a

while error >= eps:
    a = line_search(x)
    x = x-a*grad(x)
    k = k+1
    k_for_plot.append(k)
    soln.append(x)
    error = np.linalg.norm(grad(x))
    error_for_plot.append(error)
    #print(error_for_plot)

print(x)
# Plotting
plt.title("Error vs attempts")
plt.xlabel("Attempt #")
plt.ylabel("Log Error")
plt.plot(k_for_plot,error_for_plot, color = 'green')
plt.yscale('log')
plt.show()
```

Initial tested points were x2 = 0, and x3 = 0. The step size was 0.06. x2 converged to -0.1424... which is nearly -2/14. x3 converged to 0.7854... which is nearly 11/14. These values were expected after hand solving the problem.
